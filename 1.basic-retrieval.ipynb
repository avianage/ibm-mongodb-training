### Basic Retrieval

1. Display all documents in the restaurants collection.
    
    ```db.restaurants.find()```   
    
2. Display only restaurant_id, name, borough, and cuisine for all restaurants.
    
    ```db.restaurants.find({}, {restaurant_id:1 , name:1, borough:1 , cuisine:1})```

3. Same as above, but exclude _id from the results.
    
    ```db.restaurants.find({}, {_id:0,restaurant_id:1 , name:1, borough:1 , cuisine:1})```

4. Display restaurant_id, name, borough, and zipcode — again without _id.

    ```db.restaurants.find({}, {_id:0,restaurant_id:1 , name:1, borough:1 , "address.zipcode":1})```
