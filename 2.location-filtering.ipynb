### Filtering by Location & Borough

5. Find all restaurants located in the Bronx.

    ```db.restaurants.find({},{"name":1,"borough":"Bronx"})```

6. Show only the first 5 restaurants in the Bronx.

    ```db.restaurants.find({},{"name":1,"borough":"Bronx"}).limit(5)```


7. Skip the first 5 Bronx restaurants and show the next 5.

    ```db.restaurants.find({},{"name":1,"borough":"Bronx"}).skip(5).limit(5);```


10. Find restaurants with a latitude value less than -95.754168.

    ```db.restaurants.find({ "address.coord.1": { $lt: -95.754168 } })```
