### Name Pattern Matching (Regex)

14. Find restaurant_id, name, borough, and cuisine for restaurants whose name starts with 'Wil'.

    ``` db.restaurants.find({ name: /^Wil/ }, { restaurant_id: 1, name: 1, borough: 1, cuisine: 1, _id: 0 }) ```

15. Same fields for restaurants whose name ends with 'ces'.

    ``` db.restaurants.find({ name: /ces$/ }, { restaurant_id: 1, name: 1, borough: 1, cuisine: 1, _id: 0 }) ```

16. Same fields for restaurants whose name contains 'Reg' anywhere.

    ``` db.restaurants.find({ name: /Reg/ }, { restaurant_id: 1, name: 1, borough: 1, cuisine: 1, _id: 0 }) ```
