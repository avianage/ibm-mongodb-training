### Geo + Cuisine Filtering

11. Find restaurants that:
    - Do not serve American cuisine
    - Have a score greater than 70
    - Are located at a latitude less than -65.754168

    ```db.restaurants.find({$and: [{"name":1},{ "cuisine": { $ne: "American " } },{ "grades.score": { $gt: 70 } }, { "address.coord.0": { $lt: -65.754168 } }]},{"name":1,"_id":0})```

12. Same as Q11, but without using the $and operator.

    ```db.restaurants.find({"cuisine": {$ne: "American"}, "grades.score": {$gt: 70}, "address.coord.0": {$lt:-65.754168}},{"name":1,"_id":0})```

17. Find restaurants in the Bronx that serve either American or Chinese cuisine.

    ```db.restaurants.find({"borough":"Bronx","cuisine":{$in: ["American", "Chinese"]}},{"name":1, "_id":0})```
