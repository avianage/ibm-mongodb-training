### Score-Based Filtering

8. Find restaurants that achieved a score greater than 90.

    ``` db.restaurants.find({ grades: { $elemMatch: { "score": { $gt: 90 } } } }); ```

9. Find restaurants with a score between 80 and 100 (exclusive).

    ``` db.restaurants.find({ "grades.score": { $gt: 80, $lt: 100 } }) ```

20. Find restaurant_id, name, borough, and cuisine for restaurants with a score not more than 10.

    ``` db.restaurants.find({ "grades.score": { $lte: 10 } }, { restaurant_id: 1, name: 1, borough: 1, cuisine: 1, _id: 0 }) ```

30. Find restaurant_id, name, and grades for restaurants where the score is divisible by 7 (remainder = 0).

    ``` db.restaurants.find({ "grades.score": { $mod: [7, 0] } }, { restaurant_id: 1, name: 1, grades: 1, _id: 0 }) ```
